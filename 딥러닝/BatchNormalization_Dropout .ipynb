# 라이브러리 호출

```python
import tensorflow as tf
from tensorflow import keras

import numpy as np
```

# 데이터 불러오기

```python
from sklearn.datasets import load_iris
iris = load_iris()

x = iris.data
y = iris.target
```

# 데이터 분리하기

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=2021)
```

```python
train_x.shape, test_x.shape, train_y.shape, test_y.shape
```

# One Hot Encoding

```python
from tensorflow.keras.utils import to_categorical

train_y = to_categorical(train_y,3)
test_y = to_categorical(test_y,3)
```

# 모델링

```python
#레이어 선언
il = keras.layers.Input(shape=(4,))
h1 = keras.layers.Dense(256,'swish')(il)
bl = keras.layers.BatchNormalization()(h1)
dl = keras.layers.Dropout(0.2)(bl)
ol = keras.layers.Dense(3,'softmax')(dl)

#모델 선언
model = keras.models.Model(il, ol)

#모델 컴파일
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer='adam',
              metrics=['accuracy'])

#모델 확인
model.summary()

```

```python
#모델 학습
model.fit(train_x,train_y,
          verbose=1,epochs=10)
```
