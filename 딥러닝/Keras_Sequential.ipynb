# Keras Sequential API

## Linear Regression


### 라이브러리 호출

```python
import tensorflow as tf
from tensorflow import keras

import numpy as np
```

### 데이터 생성

```python
y_lst = [x*2 + 1 for x in range(0,11)]

x = np.array(range(0,11))
y = np.array(y_lst)
```

```python
print(x)
print(y)

print('shape of x : ',x.shape)
print('shape of y : ',y.shape)
```

### 모델링

```python
# 케라스 세션 클리어
keras.backend.clear_session()

#모델 선언
model = keras.models.Sequential()

#레이어 추가
model.add(keras.layers.Input(shape=(1,)))
model.add(keras.layers.Dense(1))

#모델 컴파일
model.compile(loss = 'mse' , optimizer='adam')

#모델 학습
model.fit(x,y,epochs=10,verbose=1)
```

```python
print(model.predict(x).reshape(-1))
print(y)
```

## Boston 데이터로 모델 학습

### 데이터 불러오기

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
x = boston.data
y = boston.target

x.shape, y.shape
```

### 모델링

```python
# 케라스 세션 클리어
keras.backend.clear_session()

#모델 선언
model_boston = keras.models.Sequential()

#레이어 추가
model_boston.add(keras.layers.Input(shape=(13,)))
model_boston.add(keras.layers.Dense(1))

#모델 컴파일
model_boston.compile(loss='mse', optimizer='adam')

#모델 학습
model_boston.fit(x,y,epochs=10,verbose=1)
```

```python
print(model_boston.predict(x).reshape(-1)[:5])
print(y[:5])
```

## Logistic Regression : 유방암 데이터

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
x = cancer.data
y = cancer.target

x.shape, y.shape
```

```python
# 케라스 세션 클리어
keras.backend.clear_session()

#모델 선언
model_cancer = keras.models.Sequential()

#레이어 추가
model_cancer.add(keras.layers.Input(shape=(30,)))
model_cancer.add(keras.layers.Dense(1,activation='sigmoid'))

#모델 컴파일
model_cancer.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

#모델 학습
model_cancer.fit(x,y,epochs=10,verbose=1)
```

```python
print(model_cancer.predict(x).reshape(-1)[:10])
print(y[:10])
```
